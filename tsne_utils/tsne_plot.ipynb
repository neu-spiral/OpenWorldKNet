```python

import numpy as np
import pickle
```

```python
# loading data
with open("x_new_latent_be.pkl", 'rb') as f:
    x_new_be = pickle.load(f)
with open("x_old_latent_be.pkl", 'rb') as f:  
    x_old_be = pickle.load(f)
with open("x_new_latent_ae.pkl", 'rb') as f:
    x_new_ae = pickle.load(f)
with open("x_old_latent_ae.pkl", 'rb') as f:  
    x_old_ae = pickle.load(f)
with open("y_new.pkl", 'rb') as f:
    y_new = pickle.load(f)
with open("y_old.pkl", 'rb') as f:
    y_old = pickle.load(f)
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_iter=300)

```

```python
tsne_results_old_be = tsne.fit_transform(x_old_be)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# Create the figure
fig = plt.figure( figsize=(8,8) )
ax = fig.add_subplot(1, 1, 1, title='TSNE' )
# Create the scatter
ax.scatter(
    x=tsne_results_old_be[:,0], 
    y=tsne_results_old_be[:,1], 
    c=y_old, 
    cmap=plt.cm.get_cmap('Paired'), 
    alpha=0.4)
plt.show()
```

```python
tsne_results_ae = tsne.fit_transform(np.concatenate((x_new_ae, x_old_ae)))
```

```python
fig = plt.figure( figsize=(10, 10) )
ax = fig.add_subplot(1, 1, 1, title='TSNE' )
# Create the scatter
scatter = ax.scatter(
    x=tsne_results_ae[:,0], 
    y=tsne_results_ae[:,1], 
    c=np.concatenate((y_new, y_old)), 
    cmap=plt.cm.get_cmap('Paired'), 
    alpha=0.4)
legend1 = ax.legend(*scatter.legend_elements(),
                    loc="lower left", title="Classes")
plt.show()
```
